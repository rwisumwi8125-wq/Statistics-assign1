1.Descriptive vs Inferential statistics
- Descriptive statistics: Summarizes and describes the data we already have.It does not make any predictions.

Example: Finding average marks of class 10 students in our school. Tools: Mean, Median, Mode, Std Dev, Charts.

Inferential Statistics: Uses a sample of data to make predictions or conclusion about a larger population.

Key Differences: Descriptive= "What the data shows". Inferential= "What the data predicts".

2.Sampling + Random vs stratified
- Sampling: The process of selecting a subset of individuals from a population to represent the whole population.
- Random sampling: Every member of the population has an equal chace of being selected. No groups are considered.
- Stratified sampling: Population is devided into groups/strata like age, gender. Then random samples are taken from each group proportionally.

Example: From 1000 students 500 boys and 500 girls. Take 25 boys and 25 girls randomly.

Difference: Random ignores groups. Stratified ensures all groups are represented.

3.Mean, Median, Mode
- Mean: Average = Sum of all values/Number of values
- Median: Middle most value when data is sorted. If even, average of 2 middle values.
- Mode: Most frequently occuring value.

Why important: They are "Measures of central tendency". They tell us the centre point of data.
Mean is affeted by outliers, Median is better for skewed data, Mode is useful for categorical data.

4.Skewness and Kurtosis

- Skewness: Measures asymmetry of data distribution.

Positive skew: Right tail is longer. Most ata on left. Mean> Median.

Negative skew: Left tail is longer. most data on right. Mean< Median.

Zero skew: Symmetrical. Mean = Median.

- Kustosis: Measures "tailedness" or peaknessof distribution.

High Kurtosis: Sharp peak, heavy tails. Low Kurtosis: Flat peak, light tails.

Positive skew implies: Most values are small, but there are a few very large values pulling the mean to the right. Ex: Income data.



```python
# 5: Pytho for Mean, Median, Mode

import statistics
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 26, 28]

mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

```python
# 6: Covariance and Correlation

import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 50]

cov = np.cov(list_x, list_y, ddof=0)
corr = np.corrcoef(list_x, list_y)

print("Covariance Matrix:\n", cov)
print("Correlation Matrix:\n", corr)
```

```python
# 7: Boxplot and Outliers

import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.show()

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
outliers = [x for x in data if x < lower or x > upper]
print("Outliers:", outliers)
```

```python
# 8: Ad Spend vs Sales Correlation

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

cov = np.cov(advertising_spend, daily_sales)
corr = np.corrcoef(advertising_spend, daily_sales)

print("Covariane:", cov[0,1])
print("Correlation:", corr[0,1])
```

```python
# 9: Histogram for Survey Data

import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=range(1, 12), edgecolor='black')
plt.xlabel("Satisfaction Score")
plt.ylabel("Frequency")
plt.title("Customer Satisfaction Histogram")
plt.show()

print("Mean:", np.mean(survey_scores))
print("Std Dev", np.std(survey_scores))
```
